# EXAMPLE INPUTS

```python
stdev = 0.4
corr = 0.25
```

# FIGURE

```python
import numpy as np
import plotly.graph_objects as go

num_assets = np.arange(1, 201)
variances = stdev**2 / num_assets + (num_assets-1)*stdev**2*corr / num_assets
string = "Number of Assets = %{x}<br>Standard Deviation = %{y:0.2%} <extra></extra>"
trace = go.Scatter(
    x=num_assets,
    y=np.sqrt(variances),
    mode="lines",
    hovertemplate=string,
)

fig = go.Figure(trace)
fig.update_layout(
    xaxis_title="Number of Assets",
    yaxis_title="Standard Deviation",
    yaxis_tickformat=",.0%",
    yaxis_rangemode="tozero",
    template="plotly_white"
)
fig.show()
```
