```python
import numpy as np

Num = 200       # maximum number of assets
std = 0.4       # standard deviation of each asset
cor = 0.2       # correlation between each asset
var = std ** 2  # variance
cov = var * cor # covariance

# The portfolio standard deviation decreases as the number of assets increases
stdevs = [np.sqrt(var / n + (n - 1) * cov / n) for n in range(1, Num + 1)]
stdevs
```
